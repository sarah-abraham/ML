Q1 Linear regression life time model

```python
#Exercise 1
import numpy as np
temperature_t = np.array([10,20,30,40,50,60,70,80,90]).reshape(-1, 1)
lifetime_y = np.array([420,365,285,220,176,117,69,34,5])
```

```python
import matplotlib.pyplot as plt
plt.scatter(temperature_t, lifetime_y)
plt.xlabel('Temperature')
plt.ylabel('Life time')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
lr_model = LinearRegression()
lr_model.fit(temperature_t, lifetime_y)
slope = lr_model.coef_[0]
print("b0: "+str(slope))
intercept = lr_model.intercept_
print("b1: "+str(intercept))
y_pred = lr_model.predict(temperature_t)
n = len(temperature_t)
se = np.sqrt(np.sum((y_pred - lifetime_y)**2) / (n - 2)) / np.sqrt(np.sum((temperature_t - np.mean(temperature_t))**2))
t_crit = 1.96
confidence_interval = [slope - t_crit * se, slope + t_crit * se]
print("Confidence:", confidence_interval)
```

```python
plt.scatter(temperature_t, lifetime_y)
plt.xlabel('Temperature')
plt.ylabel('Life time')
plt.plot(temperature_t,lr_model.predict(temperature_t),color='red')
plt.xlabel('Temperature')
plt.ylabel('Lifetime')
```

Q3 The values of y and their corresponding values are given, find the least sqaure regression line and value of y when x=10

```python
#Exercise 3
import numpy as np
from sklearn.linear_model import LinearRegression
x = np.array([0, 1, 2, 3, 4]).reshape(-1, 1)
y = np.array([2, 3, 5, 4, 6])
n = len(x)
mean_x = np.mean(x)
mean_y = np.mean(y)
numerator = np.sum((x - mean_x) * (y - mean_y))
denominator = np.sum((x - mean_x) ** 2)
a = numerator / denominator
b = mean_y - a * mean_x
print("y = " + str(a) + "x + " + str(b))
y_pred = a*10+b;
print("The value of y when x=10 is: "+str(y_pred))
```

```python
lr_model.fit(x, y)
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x,lr_model.predict(x),color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
