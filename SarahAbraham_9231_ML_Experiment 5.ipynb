Q1 Find the equation of hyperplane for classifying classes

```python
#Exercise 1
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
positive_data = np.array([[3, 1], [3, -1], [6, 1], [6, -1]])
negative_data = np.array([[1, 0], [0, 1], [0, -1], [-1, 0]])
X = np.vstack((positive_data, negative_data))
y = np.hstack((np.ones(len(positive_data)), np.zeros(len(negative_data))))
svm_model = SVC(kernel='linear')
svm_model.fit(X, y)
coefficients = svm_model.coef_[0]
intercept = svm_model.intercept_
equation = f"{coefficients[0]}*x + {coefficients[1]}*y + {intercept} = 0"
print("Equation of the hyperplane:", equation)
```

```python
x_min, x_max = -3, 3
y_min, y_max = -3, 3
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                     np.arange(y_min, y_max, 0.01))
Z = coefficients[0] * xx + coefficients[1] * yy + intercept
plt.figure(figsize=(10, 6))
plt.scatter(positive_data[:, 0], positive_data[:, 1], c='green', label='Positive', marker='o')
plt.scatter(negative_data[:, 0], negative_data[:, 1], c='blue', label='Negative', marker='x')
plt.contour(xx, yy, Z, colors='k', levels=[0], alpha=0.5, linestyles=['-'])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Hyperplane and Data Points')
plt.legend()
plt.grid()
plt.show()
```

Q3 Construct the weight vector of the maximum margin hyperplane

```python
#Exercise 3
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
X1 = np.array([1, 2, 2, 0, 1, 0])
X2 = np.array([1, 2, 0, 0, 0, 1])
y = np.array([1, 1, 1, -1, -1, -1])
X = np.vstack((X1, X2)).T
svm_model = SVC(kernel='linear')
svm_model.fit(X, y)
weight_vector = svm_model.coef_[0]
intercept = svm_model.intercept_[0]
print("Coefficients of the hyperplane:", weight_vector)
print("Intercept:", intercept)
```

```python
xx = np.linspace(np.min(X[:, 0]) - 1, np.max(X[:, 0]) + 1, 100)
yy = (-weight_vector[0] * xx - intercept) / weight_vector[1]
plt.scatter(X1[y == 1], X2[y == 1], c='green', marker='o', label='Positive Class')
plt.scatter(X1[y == -1], X2[y == -1], c='blue', marker='x', label='Negative Class')
plt.plot(xx, yy, 'k-')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.xlim(min(X1) - 1, max(X1) + 1)
plt.ylim(min(X2) - 1, max(X2) + 1)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid()
plt.show()
```

Q1 Find the equation of hyperplane for the dataset

```python
#Lab Assignment
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
pos = np.array([[2, 2], [3, 3], [4, 4], [5, 5], [4, 6], [3, 7], [4, 8], [5, 9], [6, 10]])
neg = np.array([[6, 2], [7, 3], [8, 4], [9, 5], [8, 8], [7, 7], [7, 8], [7, 9], [8, 10]])
X = np.vstack((pos, neg))
y = np.hstack((np.ones(pos.shape[0]), -np.ones(neg.shape[0])))
model = SVC(kernel='linear')
model.fit(X, y)
w = model.coef_[0]
b = model.intercept_[0]
equation_of_hyperplane = f"{w[0]}*x + {w[1]}*y + {b} = 0"
print("Equation of the hyperplane:", equation_of_hyperplane)
```

```python
plt.figure(figsize=(10, 8))
plt.grid(True, linestyle='--', alpha=0.7)
plt.scatter(neg[:, 0], neg[:, 1], color='blue', marker='x', label='Negative')
plt.scatter(pos[:, 0], pos[:, 1], color='green', marker='o', label='Positive')
x_vals = np.linspace(2, 9, 100)
y_vals = (-w[0] * x_vals - b) / w[1]
plt.plot(x_vals, y_vals, color='black', linewidth=2, label='Hyperplane')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Dataset and Hyperplane')
plt.show()
```
